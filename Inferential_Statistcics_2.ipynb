### **Two-sample mean test**

This test is of two types:
  * paired
  * unpaired

Two-sample mean test - paired is used when your sample observations are from the same individual or object. During this test, you are testing the same subject twice. For example, if you are testing a new drug, you would need to compare the sample before and after the drug is taken to see if the results are different.

Two-sample mean test - unpaired is used when your sample observations are independent. During this test, you are not testing the same subject twice. For example, if you are testing a new drug, you would compare its effectiveness to that of the standard available drug. So, you would take a sample of patients who consumed the new drug and compare it with another sample who consumed the standard drug.



### **Two-sample proportion test**

A two-sample proportion test, also known as a two-sample test for proportions, is a statistical test used to compare the proportions of two independent groups or populations to determine if there is a statistically significant difference between them. This test is typically used when you want to compare the proportions of a binary categorical variable (e.g., success/failure, yes/no) in two different groups. Here's how and when to perform a two-sample proportion test:

**Assumptions:**
Before conducting a two-sample proportion test, it's essential to ensure that the following assumptions are met:

1. **Independence:** The two samples are independent of each other.
2. **Random Sampling:** The samples are obtained through random sampling or an equivalent process.
3. **Sample Size:** The sample sizes should be large enough. A common rule of thumb is to have at least 5 successes and 5 failures in each sample.

**Steps to Perform a Two-Sample Proportion Test:**

1. **Formulate Hypotheses:**
   - Null Hypothesis (H0): There is no significant difference between the proportions of the two groups.
   - Alternative Hypothesis (Ha): There is a significant difference between the proportions of the two groups.

2. **Collect Data:**
   - Obtain data from your two independent groups, making sure you have the sample sizes and the counts of successes and failures in each group.

3. **Calculate Sample Proportions:**
   - Calculate the sample proportions (p1 and p2) for each group by dividing the number of successes by the total sample size in each group.

4. **Calculate the Test Statistic:**
   - The test statistic for a two-sample proportion test follows a z-distribution and is calculated as:
     ```
     z = (p1 - p2) / sqrt((p * (1 - p)) * ((1/n1) + (1/n2)))
     ```
     - p is the pooled proportion, calculated as `(x1 + x2) / (n1 + n2)` where x1 and x2 are the total successes in each group, and n1 and n2 are the sample sizes.
     - The standard error of the difference between proportions is in the denominator.

5. **Determine the Critical Value or p-value:**
   - Use the standard normal distribution table or a statistical calculator/software to find the critical value for your desired level of significance (alpha) or calculate the p-value associated with the test statistic.

6. **Make a Decision:**
   - If the test statistic falls in the rejection region (i.e., the critical value is less than the calculated test statistic, or the p-value is less than alpha), reject the null hypothesis.
   - If the test statistic does not fall in the rejection region, fail to reject the null hypothesis.

7. **Draw a Conclusion:**
   - Based on your decision, make a conclusion regarding whether there is a statistically significant difference in proportions between the two groups.

8. **Interpret the Results:**
   - If you reject the null hypothesis, interpret the results in the context of your study and consider the practical significance of the difference.

**When to Use a Two-Sample Proportion Test:**
Use a two-sample proportion test when you want to compare the proportions of a binary categorical variable in two independent groups or populations. Some common scenarios include:

- Comparing the conversion rates of two different website designs.
- Analyzing the effectiveness of two different marketing strategies (e.g., click-through rates).
- Investigating the difference in the success rates of two medical treatments.
- Comparing the proportions of defective products produced by two different machines.

In summary, a two-sample proportion test helps you determine whether there is a statistically significant difference in proportions between two groups, allowing you to make informed decisions and draw conclusions based on the data.

## **A/B Testing**

A/B testing (also known as split testing or bucket testing) is a methodology for comparing two versions of a webpage or app against each other to determine which one performs better. A/B testing is essentially an experiment where two or more variants of a page are shown to users at random, and statistical analysis is used to determine which variation performs better for a given conversion goal.
Running an A/B test that directly compares a variation against a current experience lets you ask focused questions about changes to your website or app and then collect data about the impact of that change.

Testing takes the guesswork out of website optimization and enables data-informed decisions that shift business conversations from "we think" to "we know." By measuring the impact that changes have on your metrics, you can ensure that every change produces positive results.

#### **How A/B testing works**
In an A/B test, you take a webpage or app screen and modify it to create a second version of the same page. This change can be as simple as a single headline, button or be a complete redesign of the page. Then, half of your traffic is shown the original version of the page (known as control or A) and half are shown the modified version of the page (the variation or B).

As visitors are served either the control or variation, their engagement with each experience is measured and collected in a dashboard and analyzed through a statistical engine. You can then determine whether changing the experience (variation or B) had a positive, negative or neutral effect against the baseline (control or A).



## **Chi-square test**

 Let's go through a case study using the chi-square test to analyze the relationship between two categorical variables. In this case, we will explore the association between gender and the preference for a particular type of beverage (tea, coffee, or soda) among a group of 200 people surveyed at a cafe.

### Case Study: Analyzing Beverage Preferences by Gender

#### Objective:
To determine if there is a significant association between the gender of cafe visitors and their preferences for tea, coffee, or soda.

#### Data Collection:
A cafe conducted a survey among 200 customers. The survey asked customers to select their preferred beverage from three options: tea, coffee, or soda. The cafe also recorded the gender (male or female) of each customer.

#### Hypotheses:
- Null Hypothesis (H0): Gender and beverage preference are independent (i.e., there is no significant association).
- Alternative Hypothesis (Ha): Gender and beverage preference are dependent (i.e., there is a significant association).

#### Data Presentation:
To analyze the data, we create a contingency table that shows the counts of individuals in each gender and beverage preference category:

```
                      | Tea   | Coffee | Soda  | Total
----------------------------------------------------
Male                  | 30    | 40     | 20    | 90
Female                | 20    | 45     | 45    | 110
----------------------------------------------------
Total                 | 50    | 85     | 65    | 200
```

#### Calculation:

1. Calculate the expected frequencies for each cell under the assumption of independence. The expected frequency for each cell is given by:

   Expected Frequency = (Row Total * Column Total) / Grand Total

   For example, the expected frequency for the Male-Tea cell is (90 * 50) / 200 = 22.5.

2. Calculate the chi-square statistic for each cell:

   Chi-Square = [(Observed Frequency - Expected Frequency)Â²] / Expected Frequency

   For example, for the Male-Tea cell:

   Chi-Square = [(30 - 22.5)Â²] / 22.5 â‰ˆ 12.89

3. Sum up the chi-square values for all cells to obtain the chi-square statistic for the entire table:

   Chi-Square Total â‰ˆ 28.72

4. Determine the degrees of freedom (df) for the chi-square test:

   df = (Number of Rows - 1) * (Number of Columns - 1) = (2 - 1) * (3 - 1) = 2

5. Set the significance level (Î±) for the test (e.g., Î± = 0.05).

6. Find the critical chi-square value from the chi-square distribution table for df = 2 and Î± = 0.05. The critical value is approximately 5.99.

7. Compare the calculated chi-square statistic (28.72) with the critical value (5.99). Since 28.72 > 5.99, we reject the null hypothesis.

## **F-Test or ANOVA (ANalysis Of VAriance)**

**Analysis of Variance (ANOVA)** is a statistical technique used to compare means or variations among multiple groups or treatments in a data set. It helps determine if there are statistically significant differences between these groups. ANOVA is particularly useful when you want to compare more than two groups, as it provides a way to examine group differences while controlling for the overall variability in the data. Let's dive into the details of ANOVA with an example:

### Understanding ANOVA:

#### Key Concepts:

1. **Factor**: The variable that you are investigating, often referred to as the independent variable or the treatment variable.

2. **Levels or Groups**: The different categories or treatments within the factor.

3. **Null Hypothesis (H0)**: States that there are no significant differences among the group means.

4. **Alternative Hypothesis (Ha)**: Suggests that there are significant differences among the group means.

5. **F-statistic**: The test statistic used in ANOVA, which compares the variability between group means to the variability within groups.

6. **Degrees of Freedom**: The number of independent pieces of information in the estimation of a statistic.

### Example: One-Way ANOVA

**Scenario:**
Imagine you're a quality control manager at a factory that produces widgets. You want to determine if there are any significant differences in the mean weights of widgets produced by three different machines (Machine A, Machine B, and Machine C).

#### Hypotheses:
- Null Hypothesis (H0): There are no significant differences in the mean weights of widgets produced by the three machines.
- Alternative Hypothesis (Ha): There are significant differences in the mean weights of widgets produced by the three machines.

#### Data Collection:
You collect a random sample of widgets produced by each machine and measure their weights:

- Machine A: 25 widgets with a mean weight of 12.5 ounces.
- Machine B: 30 widgets with a mean weight of 12.0 ounces.
- Machine C: 28 widgets with a mean weight of 11.8 ounces.

#### ANOVA Steps:

1. **Calculate the Grand Mean (GM)**: Add up all the individual measurements and divide by the total number of measurements (GM = 12.22 ounces in this case).

2. **Calculate the Sum of Squares Between (SSB)**:
   - SSB measures the variability between group means.
   - SSB = Î£(ni * (mean_i - GM)Â²), where ni is the sample size for group i.

   For our example:
   - SSB = (25 * (12.5 - 12.22)Â²) + (30 * (12.0 - 12.22)Â²) + (28 * (11.8 - 12.22)Â²) â‰ˆ 2.65

3. **Calculate the Sum of Squares Within (SSW)**:
   - SSW measures the variability within each group.
   - SSW = Î£Î£(xi - mean_i)Â², where xi is the individual measurement in group i.

   For our example:
   - SSW = (Î£(xi - 12.5)Â² for Machine A) + (Î£(xi - 12.0)Â² for Machine B) + (Î£(xi - 11.8)Â² for Machine C) â‰ˆ 12.16

4. **Calculate Degrees of Freedom**:
   - df_between = k - 1 = 3 - 1 = 2 (where k is the number of groups)
   - df_within = N - k = 83 - 3 = 80 (where N is the total number of measurements)

5. **Calculate the Mean Square Between (MSB)**:
   - MSB = SSB / df_between = 2.65 / 2 â‰ˆ 1.33

6. **Calculate the Mean Square Within (MSW)**:
   - MSW = SSW / df_within = 12.16 / 80 â‰ˆ 0.152

7. **Calculate the F-statistic**:
   - F = MSB / MSW â‰ˆ 1.33 / 0.152 â‰ˆ 8.75

8. **Determine the Critical Value or p-value**:
   - Consult an F-distribution table or use statistical software to find the critical value or p-value associated with the F-statistic.

9. **Make a Decision**:
   - If the calculated F-statistic is greater than the critical value or if the p-value is less than the chosen significance level (e.g., Î± = 0.05), then reject the null hypothesis.

10. **Interpret the Results**:
    - In our example, if we reject the null hypothesis, we can conclude that there are significant differences in the mean weights of widgets produced by the three machines.

### Conclusion:

ANOVA is a powerful statistical tool for comparing multiple groups and determining if there are significant differences among them.
